```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
schools_file = "Resources/schools_complete.csv"
students_file = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
schools_df = pd.read_csv(schools_file)
students_df = pd.read_csv(students_file)

# Combine the data into a single dataset
district_df = pd.merge(students_df, schools_df, how="left", on=["school_name", "school_name"])
district_df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools (answer = 15)
nb_schools = len(district_df.school_name.unique())
nb_schools
```

```python
#Calculate the total number of students (answer = 39,170)
nb_students = district_df['Student ID'].count()
nb_students
```

```python
#Calculate the total budget (answer = $24,649,428)
district_df_by_schools = district_df.groupby('school_name')
total_budget = sum(district_df_by_schools.budget.unique())
total_budget[0]
```

```python
#Calculate the average math score (answer = 78.985371)
avg_math_score = round(district_df.math_score.mean(), 6)
print(avg_math_score)

#Calculate the average reading score (answer = 81.87784)
avg_reading_score = round(district_df.reading_score.mean(), 6)
print(avg_reading_score)

#Calculate the overall average score (average of math and reading scores - answer = 80.431606)
overall_avg_score = round((avg_math_score + avg_reading_score) / 2, 6)
print(overall_avg_score)
```

```python
#Calculate the percentage of students with a passing math score (70 or greater) (answer = 74.980853)
passed_math = district_df[district_df.math_score >= 70]
nb_passed_math = passed_math['Student ID'].count()
percent_passed_math = round(nb_passed_math / nb_students * 100, 6)
print(percent_passed_math)

#Calculate the percentage of students with a passing reading score (70 or greater) (answer = 85.805463)
passed_reading = district_df[district_df.reading_score >= 70]
nb_passed_reading = passed_reading['Student ID'].count()
percent_passed_reading = round(nb_passed_reading / nb_students * 100, 6)
print(percent_passed_reading)

#Calculate the percentage of students passing both math and reading 
passed_both = district_df[(district_df.math_score >= 70) & (district_df.reading_score >= 70)]
nb_passed_both = passed_both['Student ID'].count()
percent_passed_both = round(nb_passed_both / nb_students * 100, 6)
print(percent_passed_both)
```

```python
#Create a dataframe to hold the above results
#Optional: give the displayed data cleaner formatting
summary_district_df = pd.DataFrame({"Total Schools": nb_schools,
                                    "Total Students": nb_students,
                                    "Total Budget": total_budget,
                                    "Average Math Score": avg_math_score,
                                    "Average Reading Score": avg_reading_score,
                                    "Overall Average Score": overall_avg_score,
                                    "% Passing Math": percent_passed_math,
                                    "% Passing Reading": percent_passed_reading,
                                    "% Passing Both": percent_passed_both})

summary_district_df["Total Students"] = summary_district_df["Total Students"].map("{:,}".format)
summary_district_df["Total Budget"] = summary_district_df["Total Budget"].map("${:,.2f}".format)

district_summary_results = summary_district_df[['Total Schools','Total Students','Total Budget','Average Math Score',
                                                'Average Reading Score','Overall Average Score','% Passing Math',
                                                '% Passing Reading', '% Passing Both']]
district_summary_results
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Create new table for relevant data
schools_data = district_df[['school_name', 'type', 'size', 'budget', 'reading_score', 'math_score']]

# Create columns 'passed_math', 'passed_reading', 'passed_both' (passing score has to be 70 or greater)
schools_data["passed_math"] = schools_data[schools_data.math_score >= 70]["math_score"]
schools_data["passed_reading"] = schools_data[schools_data.reading_score >= 70]["reading_score"]
schools_data["passed_both"] = schools_data[(schools_data.math_score >= 70) & (schools_data.reading_score >= 70)]["math_score"]
 
# Create the GroupBy object based on the "school_name" column
schools_group = schools_data.groupby(["school_name"])

# Extract type, budget and size per school
school_info = schools_group['type','budget','size'].max()

# Calculate average reading and math scores per school
average_scores = schools_group["reading_score", "math_score"].mean()

# Calculate # students passing reading and # students passing math for each school
count_passing = schools_group["passed_reading", "passed_math", "passed_both"].count()
```

```python
# Create a dataframe to hold the summary results per school
summary_schools_df = pd.DataFrame({"School Type": school_info["type"],
                                   "Total Students": school_info["size"],
                                   "Total School Budget": school_info["budget"],
                                   "Per Student Budget": school_info["budget"] / school_info["size"],
                                   "Average Math Score": average_scores["math_score"],
                                   "Average Reading Score": average_scores["reading_score"],
                                   "% Passing Math": count_passing["passed_math"] / school_info["size"] * 100,
                                   "% Passing Reading": count_passing["passed_reading"] / school_info["size"] * 100,
                                   "% Passing Both": count_passing["passed_both"] / school_info["size"] * 100,
                                   "% Overall Passing": (count_passing["passed_math"] + count_passing["passed_reading"])  / school_info["size"] * 50
                                  })

# Format the table
schools_summary_results = summary_schools_df[["School Type","Total Students","Total School Budget","Per Student Budget",
                                              "Average Math Score","Average Reading Score","% Passing Math",
                                              "% Passing Reading","% Passing Both","% Overall Passing"]]
schools_summary_results["Total School Budget"] = schools_summary_results["Total School Budget"].map("${:,.2f}".format)
schools_summary_results["Per Student Budget"] = schools_summary_results["Per Student Budget"].map("${:,.2f}".format)
schools_summary_results.index.names = [' ']
schools_summary_results
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create a pandas series for each grade
math_9th_series = district_df[district_df.grade == '9th'].groupby(["school_name"]).math_score.mean()
math_10th_series = district_df[district_df.grade == '10th'].groupby(["school_name"]).math_score.mean()
math_11th_series = district_df[district_df.grade == '11th'].groupby(["school_name"]).math_score.mean()
math_12th_series = district_df[district_df.grade == '12th'].groupby(["school_name"]).math_score.mean()

# Combine the series in a dataframe
math_scores_by_grade_df = pd.DataFrame({'9th': math_9th_series,
                                        '10th': math_10th_series,
                                        '11th': math_11th_series,
                                        '12th': math_12th_series})

# Format the table
math_scores_by_grade_df['9th'] = math_scores_by_grade_df['9th'].map("{:,.2f}".format)
math_scores_by_grade_df['10th'] = math_scores_by_grade_df['10th'].map("{:,.2f}".format)
math_scores_by_grade_df['11th'] = math_scores_by_grade_df['11th'].map("{:,.2f}".format)
math_scores_by_grade_df['12th'] = math_scores_by_grade_df['12th'].map("{:,.2f}".format)

math_scores_summary_results = math_scores_by_grade_df[['9th','10th','11th','12th']]
math_scores_summary_results.index.names = [' ']
math_scores_summary_results
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Define the spending bins 
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

# Slice the data and place it into the spending bins
summary_schools_df["Spending Ranges (Per Student)"] = pd.cut(summary_schools_df["Per Student Budget"],
                                                             spending_bins, labels=group_names)
```

```python
# Create a GroupBy object based upon "Spending Ranges (Per Student)"
spending_groups = summary_schools_df.groupby("Spending Ranges (Per Student)")

# Get the average of each column wanted within the GroupBy object
results_per_spending = spending_groups[["Average Math Score", "Average Reading Score", "% Passing Math",
                                        "% Passing Reading", "% Passing Both", "% Overall Passing"]].mean()
results_per_spending
```
